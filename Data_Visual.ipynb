Q5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

s_samples=1000
x = np.random.rand(s_samples) * 100       # Feature 1
y = x + (np.random.rand(s_samples) * 10)  # Feature 2 (correlated with Feature 1)
z = x - (np.random.rand(s_samples) * 10)  # Feature 3 (also correlated with Feature 1)
w = y + (np.random.rand(s_samples) * 5)   # Feature 4 (correlated with Feature 2)
v = z - (np.random.rand(s_samples) * 5)   # Feature 5 (correlated with Feature 3)

# Create a DataFrame
df = pd.DataFrame({
    'Feature 1': x,
    'Feature 2': y,
    'Feature 3': z,
    'Feature 4': w,
    'Feature 5': v
})

correlation_matrix = df.corr()

plt.title('Correlation Matrix Heatmap')

sns.heatmap(correlation_matrix,cmap='coolwarm',annot=True)
plt.show()
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px


x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
 'X': x.flatten(),
 'Y': y.flatten(),
 'Z': z.flatten()}

df = pd.DataFrame(data)

fig = px.scatter_3d(df, x='X', y='Y', z='Z',title='3D Scatter Plot of $z = \sin(\sqrt{x^2 + y^2})$',)
fig.show()

````
